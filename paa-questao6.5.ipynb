```python
import matplotlib.pyplot as plt
```

```python
x1 = [i for i in range(-1,120001)]
y1 = [180000-1.5*i for i in range(-1,120001)]
x2 = [i for i in range(-1,240001)]
y2 = [120000-i/2 for i in range(-1,240001)]
x3 = [110000 for i in range(-1,180001)]
y3 = [i for i in range(-1,180001)]
```

```python
xdot = [0, 110000, 110000, 60000, 0]
ydot = [0,0,15000,90000,120000]
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x1,y1, 'm', label = "(II)")
plt.plot(x2,y2, 'b', label = "(I)")
plt.plot(x3,y3, 'lime', label = '(III)')
legend = plt.legend(loc = 'upper right')
plt.scatter(xdot, ydot, c='k',s=20)
plt.text(0,0,'   A')
plt.text(110000,0,' B')
plt.text(110000,15000, '   C')
plt.text(60000, 90000, '   D')
plt.text(0,120000,'   E')
plt.savefig('lucro.pdf')
```

```python
lucro = 0
for i in range(110000,0,-1):
    y1 = 120000-i/2
    y2 = 180000-1.5*i
    y = min(y1,y2)
    L = 1.6*i + 1.4*y
    if L > lucro:
        lucro = L
print(lucro)
```
